```python


df = pd.read_excel("/content/stocks.xls")
df.head()
```

```python

import os
os.makedirs("data", exist_ok=True)
os.makedirs("monthData", exist_ok=True)
```

```python
avg_close = df.groupby("Name")["close"].mean().reset_index()
avg_close.columns = ["name", "avgPrice"]

avg_close.to_csv("stock_data.csv", index=False)

print("Average closing prices saved to stock_data.csv")
avg_close.head()
```

```python
for name, group in df.groupby("Name"):
    group = group.sort_values("date").copy()
    group["pct_change"] = group["close"].pct_change() * 100
    group.to_csv(f"data/{name}.csv", index=False)
print(" Daily % change to data/{Name}.csv")
```

```python
for name, group in df.groupby("Name"):
    group = group.sort_values("date").set_index("date")

    monthly = group.resample("M").agg({
        "open": "mean",
        "high": "mean",
        "low": "mean",
        "close": "mean",
        "volume": "sum"
    })

    monthly.to_csv(f"monthData/{name}.csv")
print("Monthly data saved to monthData/{Name}.csv")
```

```python
for name, group in df.groupby("Name"):
    group = group.sort_values("date").copy()
    group["rolling_30_close"] = group["close"].rolling(window=30).mean()

    path = f"data/{name}.csv"
    if os.path.exists(path):
        existing = pd.read_csv(path, parse_dates=["date"])
        updated = pd.merge(existing, group[["date", "rolling_30_close"]], on="date", how="left")
        updated.to_csv(path, index=False)
print("30-day rolling average added to data/{Name}.csv")
```
